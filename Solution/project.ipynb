```python
# combine train and test data sets
titanic_data = [train, test] 

for dataset in titanic_data:
    dataset['Title'] = dataset['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
train['Title'].value_counts()
```

```python
#title mapping
title_mapping = {
    "Mr": 0, "Miss": 1, "Mrs": 2, "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
     "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3}
for dataset in titanic_data:
    dataset['Title']=dataset['Title'].map(title_mapping)
```

```python
train.head()
```

```python
# delete unwonted features
train.drop('Name',axis=1,inplace=True)
test.drop('Name', axis=1, inplace =True)
```

```python
train.head()
```

```python
sex_mapping = {'male':0, 'female':1}
for dataset in titanic_data:
    dataset['Sex'] = dataset['Sex'].map(sex_mapping)
```

```python
train.head()
```

```python
#fill missing value
train['Age'].fillna(train.groupby('Title')['Age'].transform('median'), inplace=True)
test['Age'].fillna(test.groupby('Title')['Age'].transform('median'), inplace=True)
```

```python
# convert categorical data into numerical
for dataset in titanic_data:
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0,
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 26), 'Age'] = 1,
    dataset.loc[(dataset['Age'] > 26) & (dataset['Age'] <= 36), 'Age'] = 2,
    dataset.loc[(dataset['Age'] > 36) & (dataset['Age'] <= 62), 'Age'] = 3,
    dataset.loc[ dataset['Age'] > 62, 'Age'] = 4
```

```python
Pclass1 = train[train['Pclass']==1]['Embarked'].value_counts()
Pclass2 = train[train['Pclass']==2]['Embarked'].value_counts()
Pclass3 = train[train['Pclass']==3]['Embarked'].value_counts()
df = pd.DataFrame([Pclass1, Pclass2, Pclass3])
df.index = ['1st class','2nd class', '3rd class']
```

```python
# fill missing value in embarked
for dataset in titanic_data:
    dataset['Embarked']= dataset['Embarked'].fillna('S')
```

```python
embarked_mapping = {'S':0, 'C':1, 'Q':2}
for dataset in titanic_data:
    dataset['Embarked'] = dataset['Embarked'].map(embarked_mapping)
```

```python
# fill missing values in fare
train['Fare'].fillna(train.groupby('Pclass')['Fare'].transform('median'), inplace=True)
test['Fare'].fillna(test.groupby('Pclass')['Fare'].transform('median'), inplace=True)

train.head()
```

```python
for dataset in titanic_data:
    dataset['Cabin'] = dataset['Cabin'].str[:1]
```

```python
Pclass1 = train[train['Pclass']==1]['Cabin'].value_counts()
Pclass2 = train[train['Pclass']==2]['Cabin'].value_counts()
Pclass3 = train[train['Pclass']==3]['Cabin'].value_counts()
df = pd.DataFrame([Pclass1, Pclass2, Pclass3])
df.index = ['1st class','2nd class', '3rd class']
```

```python
cabin_mapping =  {"A": 0, "B": 0.4, "C": 0.8, "D": 1.2, "E": 1.6, "F": 2, "G": 2.4, "T": 2.8}
for dataset in titanic_data:
     dataset['Cabin'] = dataset['Cabin'].map(cabin_mapping)
```

```python
#fill missing values 
train["Cabin"].fillna(train.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
test["Cabin"].fillna(test.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
```

```python
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
test["FamilySize"] = test["SibSp"] + test["Parch"] + 1
```

```python
family_mapping = {1: 0, 2: 0.4, 3: 0.8, 4: 1.2, 5: 1.6, 6: 2, 7: 2.4, 8: 2.8, 9: 3.2, 10: 3.6, 11: 4}
for dataset in titanic_data:
    dataset['FamilySize'] = dataset['FamilySize'].map(family_mapping)
```

```python
train.head()
```

```python
features_drop = ['Ticket', 'SibSp','Parch']
train = train.drop(features_drop, axis=1)
test = test.drop(features_drop, axis=1)
train = train.drop(['PassengerId'], axis =1)
```

```python
train_data = train.drop('Survived', axis=1)
target = train['Survived']

train_data.shape, target.shape
```

```python
train_data.head()
```

```python
#modelling

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
import numpy as np
```

```python
# cross validation

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

# Decision tree

```python
clas = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clas, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# decision tree score
round(np.mean(score)*100,2)
```

# Random forest

```python
clas = RandomForestClassifier(n_estimators = 13)
scoring = 'accuracy'
score = cross_val_score(clas, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# random forest model score
round(np.mean(score)*100,2)
```

# Naive bayes

```python
clas = GaussianNB()
scoring = 'accuracy'
score = cross_val_score(clas, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
#naive bayes score
round(np.mean(score)*100,2)
```

```python
#testing

clas = DecisionTreeClassifier()
clas.fit(train_data, target)

test_data = test.drop('PassengerId', axis=1).copy()
prediction = clas.predict(test_data)
```

```python
sub = pd.DataFrame({'PassengerId':test['PassengerId'], 'Survived': prediction})

sub.to_csv('result.csv', index=False)
```
